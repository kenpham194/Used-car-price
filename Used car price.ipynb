```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_cars=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data',header=None)
```

```python
headers= ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df_cars.columns=headers
```

```python
df_cars.head()
```

```python
df_cars.dtypes
```

```python
plt.figure(1)
plt.subplot(221)
df_cars['symboling'].plot(kind='hist',figsize=(10,6),title='symboling')
plt.subplot(222)
df_cars['length'].plot(kind='hist',title='length')
plt.subplot(223)
df_cars['height'].plot(kind='hist',title='height')
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
df_cars['city-mpg'].plot(kind='hist',figsize=(10,6),title='city-mpg')
plt.subplot(222)
df_cars['highway-mpg'].plot(kind='hist',title='highway-mpgd')
plt.show()
```
